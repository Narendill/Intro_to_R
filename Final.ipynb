```python
import matplotlib.pyplot as plt
from sympy.solvers import solve
from sympy import Symbol, sin, cos
```

# 4. Построить график
Для лучшего понимания поведения функции в первую очередь построим ее график.

```python
x = np.arange(-100, 100.1, 0.01)
y = [-12 * i**4 * sin(cos(i)) - 18 * i**3 + 5 * i**2 + 10 * i - 30 for i in x]

fig, ax = plt.subplots()
ax.set_title('График функции в промежутке [-100, 100]', fontsize=10, pad = 20)
ax.plot(x, y, linewidth=3)
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tick_params(axis='both', which='major', labelsize=6)
plt.show()
```

Из графика видно, что исследуемая функция содержит бесконечное количество корней, промежутков возрастания и убывания, а также экстремумов (*solve* из библиотеки *sympy* не может найти корни: *No algorithms are implemented to solve equation*). 

Попробуем исследовать поведение функции в конкретном диапазоне, например в промежутке $[-1; 1]$.

Для начала построим график функции в указанном диапазоне.

```python
x = np.arange(-1, 1.001, 0.001)
y = [-12 * i**4 * sin(cos(i)) - 18 * i**3 + 5 * i**2 + 10 * i - 30 for i in x]

fig, ax = plt.subplots()
ax.set_title('График функции в промежутке [-1, 1]', fontsize=10, pad = 20)
ax.plot(x, y, linewidth=3)
ax.spines['left'].set_position(('data', 0)) #('center')
ax.spines['bottom'].set_position(('data', -22)) #('center')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tick_params(axis='both', which='major', labelsize=6)
plt.show()
```

# 1. Определить корни
Из графика функции видно, что на промежутке $[-1; 1]$ у функции не существует корней, поскольку все ее значения принадлежат отрицательному диапазону, то есть в указанном интервале у функции нет таких значений, чтобы $f(x) = 0$

```python
print(f'Область значений функции: от {round(min(y), 2)} до {round(max(y), 2)}')
```

Таким образом функция не пересекает ось абсцисс в рассматриваемом диапазоне.

# 2-3. Найти интервалы, на которых функция возрастает и убывает
Определение интервалов возрастания и убывания функции также будем определять исходя из рассчитанных координат.

```python
# Определяем "индексы возрастания"
positions_vozrost = []

for i in range(len(y) - 1):
    if y[i+1] > y[i]:
        positions_vozrost.append(i)
```

```python
fig, ax = plt.subplots()
ax.set_title('Промежутки возрастания и убывания функции', fontsize=10, pad = 20)
ax.plot(x, y, linewidth=3)
ax.plot(x[positions_vozrost[0]:positions_vozrost[-1]], y[positions_vozrost[0]:positions_vozrost[-1]],
        color='red', linewidth=3)
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', -22))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tick_params(axis='both', which='major', labelsize=6)
plt.show()
```

```python
print(f'Промежуток возрастания функции: ({round(x[positions_vozrost[0]], 2)}, {round(x[positions_vozrost[-1]], 2)})')        
print(f'Промежутки убывания функции: ({round(x[0], 2)}, {round(x[positions_vozrost[0]], 2)})',\
      f'и ({round(x[positions_vozrost[-1]], 2)}, {round(x[-1], 2)})')
```

Таким образом, функция возрастает в интервале: $(-0.39, 0.45)$, а убывает в интервалах $(-1.0, -0.39)$ и $(0.45, 1.0)$.

# 5. Вычислить вершину
Из графика видно, что на рассматриваемом промежутке у функции имеются две вершины. Вычислим координаты этих точек.

```python
print(f'Первая вершина: ({round(x[positions_vozrost[0]], 2)}, {round(y[positions_vozrost[0]], 2)})')
print(f'Вторая вершина: ({round(x[positions_vozrost[-1]], 2)}, {round(y[positions_vozrost[-1]], 2)})')
```

Таким образом, вершины функции в рассматриваемом диапазоне находятся в точках $(-0.39, -32.29)$ и $(0.45, -26.51)$. Отметим их на графике:

```python
fig, ax = plt.subplots()
ax.set_title('Вершины функции', fontsize=10, pad = 20)
ax.plot(x, y, linewidth=3)
ax.plot(x[positions_vozrost[0]:positions_vozrost[-1]], y[positions_vozrost[0]:positions_vozrost[-1]],
        color='red', linewidth=3)
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', -22))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tick_params(axis='both', which='major', labelsize=6)

ax.plot(-0.39, -32.29, 'go', markersize = 10)
ax.plot(0.45, -26.51, 'go', markersize = 10)

plt.show()
```
